```
from tensorflow import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data()
word_index = keras.datasets.imdb.get_word_index()
word_to_num = {word:(num+3) for word,num in word_index.items()}
word_to_num["<PAD>"] = 0
word_to_num["<START>"] = 1
word_to_num["<UNKNOWN>"] = 2
num_to_word = {num:word for word,num in word_to_num.items()}
print(' '.join(num_to_word[num] for num in x_train[0]))
print('x_train[0] is positive' if y_train[0] == 1 else 'x_train[0] is negative')
print()
print(' '.join(num_to_word[num] for num in x_train[1]))
print('x_train[1] is positive' if y_train[1] == 1 else 'x_train[1] is negative')
```
